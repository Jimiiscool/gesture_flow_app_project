```python
import mediapipe as mp
import cv2
import numpy as np
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
mp_hands = mp.solutions.hands
```

```python
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        ret, frame = cap.read()

        #Change color to detectable format
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        #Process image
        results = hands.process(img)

        #Change color back
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

        #Print Results.
        print(results)

        if results.multi_hand_landmarks:
            cords_list = np.array([])
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(img, hand, mp_hands.HAND_CONNECTIONS)
                for i, cords in enumerate(hand.landmark):
                    #hand_cords = np.array([cords.x, cords.y])
                    x_cords = f'X Cords: {cords.x}'
                    y_cords = f'Y Cords: {cords.y}'
                    cords_list = np.append(cords_list, (x_cords, y_cords), axis=0)
        
        cv2.imshow("Hand Tracking Feed", img)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
